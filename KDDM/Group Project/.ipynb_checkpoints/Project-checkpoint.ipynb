```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Reading the CSV file

```python
dataset = pd.read_csv("data_2.csv")
dataset
```

```python
#Always return average, median. if the average is 0, then we have a mode
#mode is the median when the values are not numerical
def AverageMedian(df, dfColumn):
    if df[dfColumn].dtype in ['int64', 'float64']:
        return df[dfColumn].mean(), df[dfColumn].median()
    elif df[dfColumn].dtype == 'bool':
        mean_val = df[dfColumn].mean()
        median_val = df[dfColumn].median()
        return mean_val, median_val
    elif df[dfColumn].dtype == 'object':
        median_val = df[dfColumn].mode().iloc[0]
        return 0, median_val
    else:
        return "Unsupported data type"
    
def checkNulls(df, dfColumn):
    result = "Nulls=" + str( df[dfColumn].isnull().sum().sum() )
    return result
    
```

# Cleaning the data


```python
#Easiest is removing duplicates 
datasetNoDuplicates = dataset.drop_duplicates() #removed 4 rows
datasetNoDuplicates
```

# Script to replace NaN values with average or median. For any column


```python
#before cleaning had 329 values
SpecifiedColumn = 'Age'
average, median = AverageMedian(datasetNoDuplicates, SpecifiedColumn)
print(f"Average: {average} Median: {median}" + " " +checkNulls(datasetNoDuplicates, SpecifiedColumn) )

print("Do you want to replace NaN values in the " + SpecifiedColumn + " column with the average, median or custom?")
response = input()

#now we take care that if the user input is not correct, we loop until it is
while( response != 'average' and response != 'median' and response != 'quit' and response != "custom" ):
    print("Please type your answer in lowercase letters.")
    response = input()
    
#Once we have the correct input, we react
if response == "average":
    datasetNoDuplicates.loc[:, SpecifiedColumn] = datasetNoDuplicates[SpecifiedColumn].fillna(average) 
elif response == "median":
    datasetNoDuplicates.loc[:, SpecifiedColumn] = datasetNoDuplicates[SpecifiedColumn].fillna(median) 

#Custom value to impute
elif response == "custom":
    print("is the new custom value going to be a string or a float?")
    customType = input()
    #while input is not valid, repeat
    while(customType != "string" and customType != "float"):
        print("Please type string or float")
        customType = input()
        
    if(customType == "string"):
        print("String value. Which value should we replace NaN with?")
        customValue = input()
    elif(customType == "float"):
        print("Float value. Which value should we replace NaN with?")
        customValue = float( input() )
    
datasetNoDuplicates.loc[:, SpecifiedColumn] = datasetNoDuplicates[SpecifiedColumn].fillna(customValue)
        
print("...Done!")
```

```python
#Test if the Number of Nulls is 0
averageAfter, medianAfter = AverageMedian(datasetNoDuplicates, SpecifiedColumn)
print(f"Average: {averageAfter} Median: {medianAfter}" + " " +checkNulls(datasetNoDuplicates, SpecifiedColumn) )
```
